```python
train = train[['text', 'category']]
data_to_predict = data_to_predict[['text']]
```

```python
from sklearn.model_selection import train_test_split
TRAIN_SIZE = 0.75
dataset_count = len(train)

df_train, df_test = train_test_split(train, test_size=1-TRAIN_SIZE, random_state=42)

print("TRAIN size:", len(df_train))
print("TEST size:", len(df_test))
```

```python
import time
import os
from joblib import Parallel, delayed

train_len = len(df_train.index)
texts_train = df_train['text'].tolist()
classes_train = df_train['category'].tolist()

test_len = len(df_test.index)
texts_test = df_test['text'].tolist()
classes_test = df_test['category'].tolist()
```

```python
path = os.getcwd()
train_path = os.path.join(path, "..", "kaggle_datasets", 'FASAM', 'bert', 'train', '')
test_path = os.path.join(path, "..", "kaggle_datasets", 'FASAM', 'bert', 'test', '')

def write_txt(text_, class_, i, is_train):
     if is_train == 0:
         cur_path = test_path
     else:
         cur_path = train_path
     with open(os.path.join(cur_path, class_, f'{i}_{class_}.txt'), 'w', encoding='utf-8') as f:
         f.write(text_)

threads=4
Parallel(n_jobs=threads)(delayed(write_txt)(texts_train[i], classes_train[i], i, 1) for i in range(train_len))
Parallel(n_jobs=threads)(delayed(write_txt)(texts_test[i], classes_test[i], i, 0) for i in range(test_len))
```

```python
import ktrain
from ktrain import text
import os
import numpy as np 
import pandas as pd
```

```python
train = pd.read_csv('../kaggle_datasets/FASAM/train.csv')
test = pd.read_csv('../kaggle_datasets/FASAM/test.csv')
submission = pd.read_csv('../kaggle_datasets/FASAM/sample_submission.csv')
```

```python
train = train[['text', 'category']]
test = test[['text']]
```

```python
data_to_predict=test['text'].tolist()
data_to_predict
```

```python
os.environ['CUDA_VISIBLE_DEVICES'] = '-1' #To ignore GPU
```

```python
(x_train, y_train), (x_test, y_test), preproc = text.texts_from_folder("../kaggle_datasets/FASAM/bert",
                                                                        maxlen=500,
                                                                        preprocess_mode="bert",
                                                                        classes=["ambiente", "ciencia", "comida", "educacao", "empreendedorsocial", "equilibrioesaude", "sobretudo", "tec", "turismo"])
```

```python
learner = ktrain.get_learner(text.text_classifier("bert", (x_train, y_train), preproc=preproc),
                            train_data = (x_train, y_train),
                            val_data = (x_test, y_test),
                            batch_size = 6)
```

```python
learner.fit_onecycle(2e-5, 3)
```

```python
predictor = ktrain.get_predictor(learner.model, preproc)
predictions = predictor.predict(data_to_predict)
```

```python
submission['category'] = predictions
submission.to_csv('submission_bert_portuguese.csv', index=False)
```
